# NumPy Operations

## Arithmetic

You can easily perform array with array arithmetic, or scalar with array arithmetic. Let's see some examples:

```python
import numpy as np

myArray = np.arange(0,10)

# show array
myArray
```

```python
# Performing addition operation
myArray + myArray
```

```python
# Performing subtraction operation
myArray - myArray
```

```python
# Warning on division by zero, but not an error!
# Just replaced with nan

myArray / myArray
```

```python
# Also warning, but not an error instead infinity

1/myArray
```

```python
# Performing exponentiation
myArray**2
```

```python
li = [34,56,78]

for i,j in zip(li,li):
    print(i+j)
    
```

## Universal Array Functions

Numpy comes with many [universal array functions](http://docs.scipy.org/doc/numpy/reference/ufuncs.html), which are essentially just mathematical operations you can use to perform the operation across the array. Let's show some common ones:

```python
#Taking Square Roots
np.sqrt(myArray)
```

```python
#Calcualting exponential (e^)
np.exp(myArray)
```

```python
#same as arr.max()
np.max(myArray)    
```

```python
np.sin(myArray)
```

```python
np.log(myArray)
```

## Splitting Arrays

```python
myArray = np.arange(1,25).reshape((6,4))
myArray
```

```python
# Splitting horizontally
np.hsplit(myArray,2)
```

```python
# Splitting vertically 

np.vsplit(myArray,3)
```

```python
# Unpacking of a list

arr1,arr2,arr3 = np.vsplit(myArray,3)
print(arr1)
print(arr2)
print(arr3)
```

## Joining Arrays

```python
# Joining arrays by their rows

combined_rows = np.concatenate([arr1,arr2,arr3])
combined_rows
```

```python
# Joining arrays by their columns

combined_cols = np.concatenate([arr1,arr2,arr3], axis=1)
combined_cols
```
