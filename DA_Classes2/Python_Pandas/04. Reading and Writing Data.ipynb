```python
# Reading data from a text file without headers

df = pd.read_table('data2.txt', 
                   sep=',',
                   header=None,
                  names=['Name','Score','Gender'])
df
```

```python
# Reading data from a text file without headers

df = pd.read_csv('data2.txt', 
                   header=None,
                  names=['Name','Score','Gender'])
df
```

```python
# Reading data from a text file without headers

df = pd.read_csv('data7.txt', 
                   header=None,
                   delimiter="$",
                  names=['Name','Score','Gender'])
df
```

```python
# Reading data from a text file having extra information at header

df = pd.read_csv('data4.txt', skiprows=2)
df
```

```python
# Reading data from a text file having missing values

df = pd.read_csv('data5.txt', na_values='n.a')
df
```
