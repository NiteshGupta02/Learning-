# DataFrame

## What is DataFrame?

```python
import pandas as pd
```

```python
# Creating a dataframe using a dictionary

mydict = {
    'name':['james','roger'],
    'age':[40,35],
    'sex':['male','male'],
    'country':['sweden','us']
}
print(mydict)
df = pd.DataFrame(mydict)
df
```

```python
mydict2 = {
    'Name':['Amit','Hina','Rahul','Himani'],
    'Age':[34,24,28,25],
    'Sex':['Male','Female','Male','Female'],
    'Address':['Delhi','Mumbai','Indore','Bhopal']
}
df1 = pd.DataFrame(mydict2)
df1
```

```python
# Creating a dataframe from series without column names

S1 = pd.Series(['Python','Data Science','C'])
S2 = pd.Series([20000,35000,12000])
S3 = pd.Series([2000,3000,1000])

df = pd.concat([S1,S2,S3],axis=1)
df
```

```python
# Creating a dataframe from series with column names

S1 = pd.Series(['Python','Data Science','C'], name='Course')
S2 = pd.Series([20000,35000,12000], name='Fee')
S3 = pd.Series([2000,3000,1000], name='Discount')

df = pd.concat([S1,S2,S3],axis=1)
df
```

```python
# Creating a dataframe using list

course = ['Python','Data Science','C']
fee = [20000,35000,12000]
discount = [2000,3000,1000]

df = pd.DataFrame(zip(course,fee,discount),
                  columns=['Course','Fee','Discount'])
df
```

```python
# Create a dictionary using list of tuples

myDict = dict([('Python',20000),
               ('Data Science',45000), 
               ('C',12000)])

print(myDict)
```

```python
df = pd.DataFrame(dict([('Python',20000),
               ('Data Science',45000), 
               ('C',12000)]), index=[0])
df
```

```python
# Create a dictionary using tuple of lists

df = pd.DataFrame(dict((['Python',20000],
               ['Data Science',45000], 
               ['C',12000])), index=[0])
df
```
