## What is Decision Tree?

Decision tree is a supervised learning algorithm which is used for both classification and regression. As the name goes, it uses a tree-like model of decisions, where each internal node denotes a test on an attribute, each branch represents an outcome of the test, and each leaf node (terminal node) holds a class label.

# Terminology:

- **Root Node:** The topmost node that represents the entire dataset. It has no incoming, only outgoing edges.

- **Internal Nodes (Decision Nodes):** These are the non-leaf nodes in a decision tree. They represent the splitting of the dataset into subsets based on the values of the feature represented by the node. These have both incoming and outgoing edges.

- **Leaf Nodes (Terminal Nodes):** A decision tree's end nodes. They represent the final classification of a subset of the dataset. These have only incoming edges.

- **Max Depth:** The layers count between the root and last leaf nodes. (In the above example, figure 1, max depth is 2). 

- **Threshold Values:** The numbers, which are used to split the dataset of each decision nodes.
